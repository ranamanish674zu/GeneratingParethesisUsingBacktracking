```python
#backtracking generate paranthesis
def genParenthesis(openB,closeB,n,s=[]):
    #Base Case 
    if(closeB==n):
        print(''.join(s))
        return 
    else :
        if(openB>closeB):
            s.append(')')
            genParenthesis(openB,closeB+1,n,s)
            s.pop()
        if(openB<n):
            s.append('( ')
            genParenthesis(openB+1,closeB,n,s)
            s.pop()
    return 
n=int (input("enter the no of paranthesis\n"))
genParenthesis(1,1,n)
```

```python
#Iteration which use to  an iterator is an object that allows you to traverse through a collection of elements one by one.
# Iterators are used to iterate over sequences such as lists, tuples, dictionaries, and sets.
numbers = [1, 2, 3, 4, 5]
it = iter(numbers)

print(next(it))  # prints 1
print(next(it))  # prints 2
print(next(it))  # prints 3
print(next(it))  # prints 4
print(next(it))  # prints 5
```

```python
#a generator is a special type of iterator that allows you to generate a sequence of values on-the-fly, rather than having to
# create the entire sequence in memory upfront.
def square_generator(n):
    for i in range(n):
        yield i**2
for square in square_generator(5):
    print(square)
```
